```python
import sys
import itertools
import pandas as pd
import plotly.express as px
sys.path.append('..') 
import seaborn as sns

from model.run import df  
df1 = df.copy(deep=True)

# df1 = pd.read_csv("20_000_to_100_8.csv")

my_colors = ["#253659", "#BF665E", "#F28157"] 
```

```python
df1 = df.copy(deep=True)
#df2 = df1.groupby(['timestep']).last().reset_index()
df1
```

```python
df2 = df1.copy(deep=True)
df2 = df2.groupby(['run', 'timestep']).last().reset_index()

fig = px.line(df2, x="timestep", y=[ 'new_view'], line_group='run',  title='Количество стартовых просмотров', height=600, color_discrete_sequence= my_colors)


fig.update_layout(xaxis_title='Месяцы', yaxis_title='Просмотры')

fig.show()
```

```python
df2 = df1.copy(deep=True)
df2 = df2.groupby(['run', 'timestep']).last().reset_index()

fig = px.line(df2, x="timestep", y=[ 'count_view','buks'], line_group='run',  title='Просмотры в месяц', height=600, color_discrete_sequence= my_colors)
# Установите логарифмическую шкалу для оси Y
fig.update_yaxes(type='log')


fig.update_layout(xaxis_title='Месяцы', yaxis_title='Просмотры')

fig.show()
```

```python
df2 = df1.copy(deep=True)
df2 = df2.groupby(['run', 'timestep']).last().reset_index()

fig = px.line(df2, x="timestep", y=[ 'buks'], line_group='run',  title='Доходы в месяц', height=600, color_discrete_sequence= my_colors)


fig.update_layout(xaxis_title='Месяцы', yaxis_title='Доход')

fig.show()
```

```python
df2 = df1.copy(deep=True)
df2 = df2.groupby(['run', 'timestep']).last().reset_index()

fig = px.bar(df2, x="timestep", y=[ 'count_view','new_view'],   title='Просмотры в месяц', height=600, color_discrete_sequence= my_colors)


fig.update_layout(xaxis_title='Месяцы', yaxis_title='Просмотры')
fig.update_layout(barmode='group')

fig.show()
```

```python
df2 = df1.copy(deep=True)
df2 = df2.groupby(['run', 'timestep']).last().reset_index()

fig = px.line(df2, x="timestep", y=[ 'formula'], line_group='run',  title='Скорость продолжительности набора просмотров', height=500, color_discrete_sequence= my_colors)


fig.update_layout(xaxis_title='Месяцы', yaxis_title='Просмотры')

fig.show()
```
